```python
words = open('names.txt', 'r').read().splitlines()
words[:3]
```

```python
len(words)
```

```python
## BIGRAM implementation

freq = {}

for w in words:
  tokens = f".{w}."
  for a, b in zip(tokens, tokens[1:]):
    bigram = (a, b) 

    freq[bigram] = freq.get(bigram, 0) + 1

    # print(a, b)

sorted(freq.items(), key=lambda kv: -kv[1])
```

```python
import torch
```

```python
N = torch.zeros((27, 27), dtype=torch.int32)
```

```python
chars = sorted(list(set(''.join(words)))) # set of all possible unique characters

stoi = { s: i+1 for i, s in enumerate(chars) }
stoi['.'] = 0

itos = { i: s for s, i in stoi.items()} 
itos
```

```python
for w in words:
  tokens = f".{w}."
  for ch1, ch2 in zip(tokens, tokens[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    N[ix1, ix2] += 1
```
